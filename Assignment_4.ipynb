Load data using pandas dataframe

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# loading data using pandas dataframe
houses = pd.read_csv("houses_to_rent.csv")
print(houses)
```

Fraw a scatterplot using at least 2 features

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(x="rooms", y="parking spaces", data=houses)
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.show()
```

- Draw box plots of all features

```python
box_data= houses
sns.boxplot(data = box_data, width=0.5, fliersize=5)
plt.show()
```

do a heatmap analysis, mentioned 2 most correlated features. explain why

```python
cor=houses.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

bathroom and rooms are most correlated features because the value(0.75) closest to one is between these two features.
